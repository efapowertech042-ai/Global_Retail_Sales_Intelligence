```python
import pandas as pd

df = pd.read_csv("../SQL_Engineered_Dataset/ml_monthly_revenue_dataset.csv")

df.head()
```

```python
# Dataset quick checks

df.info()
df.isna().sum()
```

### Define Target and Features

```python
target = "TARGET_REVENUE_USD"

X = df.drop(columns=[target, "YEAR_AND_MONTH"])
y = df[target]
```

### Identify Feature Types

```python
categorical_features = ["PRODUCTKEY", "STOREKEY", "MONTH", "QUARTER"]
numeric_features = [
    "TOTAL_UNITS_SOLD",
    "REVENUE_LAST_MONTH",
    "REVENUE_3_LAST_MONTH"
]
```

### Preprocessing (Encoding)

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features),
        ("num", "passthrough", numeric_features)
    ]
)
```

### Time-Aware Train/Test Split (Very Important)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    shuffle=False   # ðŸš¨ do NOT shuffle time data
)
```

### Baseline Model â€” Linear Regression

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression

lr_model = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("model", LinearRegression())
])

lr_model.fit(X_train, y_train)
```

### Evaluate the Model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred = lr_model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"MAE:  {mae:,.2f}")
print(f"RMSE: {rmse:,.2f}")
print(f"RÂ²:   {r2:.3f}")
```

### Inspect Feature Importance (This Is Gold)
#### Interpretation of Product Coefficients

The product coefficients represent the average *relative* monthly revenue uplift a product contributes compared to a baseline product, after controlling for store effects, seasonality, and past revenue trends. These values help identify products that consistently drive higher revenue and therefore deserve more operational and strategic attention. Importantly, this reflects **revenue impact**, not profitability or margin.

```python
feature_names = (
    lr_model.named_steps["preprocessor"]
    .get_feature_names_out()
)

coefficients = lr_model.named_steps["model"].coef_

importance_df = (
    pd.DataFrame({
        "feature": feature_names,
        "coefficient": coefficients
    })
    .sort_values(by="coefficient", key=abs, ascending=False)
)

importance_df.head(15)
```

### Product Coefficient Visualization

```python
import matplotlib.pyplot as plt

# Filter product-related coefficients
product_coef = importance_df[
    importance_df["feature"].str.contains("cat__PRODUCTKEY")
]

# Select top revenue-driving products
top_products = product_coef.sort_values(
    by="coefficient", ascending=False
).head(10)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(
    top_products["feature"],
    top_products["coefficient"]
)
plt.axvline(0)  # baseline reference
plt.xlabel("Relative Revenue Impact (USD)")
plt.ylabel("Product")
plt.title("Top Revenue-Driving Products (monthly revenue uplift compared to baseline product)")
plt.gca().invert_yaxis()
plt.show()
```

### Actual vs Predicted Monthly Revenue

This plot compares the modelâ€™s predicted monthly revenue against the actual observed revenue. Each point represents a productâ€“storeâ€“month observation, while the diagonal line indicates perfect prediction.

An RÂ² score of **0.59** shows that the model explains approximately **59% of the variance** in monthly revenue, indicating meaningful predictive power for a real-world business dataset. The **Mean Absolute Error (MAE) of $458** suggests that, on average, predictions deviate from actual revenue by about $458 per month.

The clustering of points near lower revenue values reflects the natural distribution of sales, where most productâ€“store combinations generate modest monthly revenue. As revenue increases, prediction variance widens, indicating that high-revenue months are harder to predict precisely. This behavior is expected in revenue forecasting problems due to promotions, bulk purchases, or external demand shocks not fully captured by the model.

Overall, this visualization demonstrates that incorporating product-level effects, seasonality, store context, and lagged revenue features enables the model to produce stable and interpretable monthly revenue forecasts, while acknowledging increased uncertainty at higher revenue levels.
