### Sales Time Analysis EDA

```python
sales_time = sales_df[['ORDER_DATE', 'PRODUCTKEY', 'QUANTITY']]
products_price = products_df[['PRODUCTKEY', 'UNIT_PRICE_USD']]
```

```python
# Merge Sales + Product Price

sales_time_df = sales_time.merge(
    products_price,
    on='PRODUCTKEY',
    how='left'
)
```

```python
# Create Revenue

sales_time_df['REVENUE_USD'] = (
    sales_time_df['QUANTITY'] * sales_time_df['UNIT_PRICE_USD']
)
```

```python
# Proper Time-Based Analysis

sales_time_df['ORDER_DATE'] = pd.to_datetime(sales_time_df['ORDER_DATE'])

yearly_revenue = (
    sales_time_df
    .groupby(sales_time_df['ORDER_DATE'].dt.year)['REVENUE_USD']
    .sum()
    .reset_index(name='TOTAL_REVENUE_USD')
)

yearly_revenue
```

```python
# Visualization

import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.plot(yearly_revenue['ORDER_DATE'], yearly_revenue['TOTAL_REVENUE_USD'])
plt.title("Yearly Revenue Trend (USD)")
plt.xlabel("Year")
plt.ylabel("Total Revenue")
plt.grid(True)
plt.show()
```

### Exchange Rate Exploratory Data Analysis (EDA)

#### - Load and Inspect Exchange Rate Data

This section loads the exchange rate dataset and performs an initial inspection to understand its structure, data types, and overall completeness.  
The goal is to confirm that dates, currencies, and exchange values are correctly represented before deeper analysis.

```python
exchange_df['DATE'] = pd.to_datetime(exchange_df['DATE'])
exchange_df.info()
```

```python
# date formatting
exchange_df['DATE'] = pd.to_datetime(exchange_df['DATE'])
```

#### Currency Coverage Analysis
This step identifies all currencies present in the dataset and checks how
frequently each appears to ensure consistent historical coverage.

```python
exchange_df['CURRENCY'].value_counts()
```

#### - Date Range Assessment
We evaluate the start and end dates of the exchange rate data to ensure
alignment with the sales data timeline.

```python
exchange_df['DATE'].min(), exchange_df['DATE'].max()
```

#### - Exchange Rate Distribution and Volatility
Descriptive statistics are calculated per currency to understand central
tendencies, variability, and potential volatility risks.

```python
exchange_df.groupby('CURRENCY')['EXCHANGE'].describe()
```

#### - Exchange Rate Trends Over Time
This visualization shows how exchange rates evolve over time relative to USD,
helping identify long-term appreciation or depreciation patterns.

```python
plt.figure(figsize=(10, 5))

for currency in exchange_df['CURRENCY'].unique():
    subset = exchange_df[exchange_df['CURRENCY'] == currency]
    plt.plot(subset['DATE'], subset['EXCHANGE'], label=currency)

plt.legend()
plt.title("Exchange Rate Trends vs USD")
plt.xlabel("Date")
plt.ylabel("Exchange Rate")
plt.show()
```

#### - Daily Exchange Rate Movement
We compute daily percentage changes in exchange rates to normalize movements
and enable fair volatility comparison across currencies.

```python
exchange_df['DAILY_PCT_CHANGE'] = (
    exchange_df
    .groupby('CURRENCY')['EXCHANGE']
    .pct_change()
)
```

#### - Currency Volatility Comparison
Currencies are ranked by the standard deviation of their daily percentage
changes to identify those with higher exchange-rate risk.

```python
exchange_df.groupby('CURRENCY')['DAILY_PCT_CHANGE'].std().sort_values(ascending=False)
```
